```python
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
from skopt import BayesSearchCV
from skopt.space import Categorical, Real, Integer

# Define the parameter space for LGBMRegressor using Categorical for discrete values
param_grid = {
    'n_estimators': Categorical([100, 200, 300]),
    'learning_rate': Categorical([0.01, 0.05, 0.1]),
    'num_leaves': Categorical([20, 31, 40]),
    'max_depth': Categorical([5, 7, 10]),
    'min_child_samples': Categorical([20, 30]),
    'subsample': Categorical([0.8, 1.0]),
    'colsample_bytree': Categorical([0.8, 1.0]),
    'reg_alpha': Categorical([0.1, 0.5, 1.0]),
    'reg_lambda': Categorical([0.1, 0.5, 1.0])
}

# Initialize LGBMRegressor
lgbm = LGBMRegressor(random_state=42)

# Initialize BayesSearchCV
bayes_search = BayesSearchCV(
    estimator=lgbm,
    search_spaces=param_grid, # Use search_spaces for BayesSearchCV
    scoring='neg_mean_squared_error', # Using negative MSE for maximization
    cv=3, # 3-fold cross-validation
    n_jobs=-1, # Use all available cores
    verbose=1,
    n_iter=50 # Number of iterations for Bayesian Optimization (can be adjusted)
)

# Fit BayesSearchCV to the training data
bayes_search.fit(x_train, y_train)
```

```python
# Print the best parameters and best score
print(f"Best Parameters: {bayes_search.best_params_}")
print(f"Best Negative MSE: {bayes_search.best_score_:.2f}")

# Get the best model
best_lgbm_model = bayes_search.best_estimator_

# Make predictions on the test set with the best model
y_pred_tuned_lgbm = best_lgbm_model.predict(x_test)

# Evaluate the best model
mse_tuned = mean_squared_error(y_test, y_pred_tuned_lgbm)
mae_tuned = mean_absolute_error(y_test, y_pred_tuned_lgbm)
mape_tuned = mean_absolute_percentage_error(y_test, y_pred_tuned_lgbm)
r2_tuned = r2_score(y_test, y_pred_tuned_lgbm)

print(f"\nFine-tuned LGBM Mean square Error: {mse_tuned:.2f}")
print(f"Fine-tuned LGBM R2: {r2_tuned:.2f}")
print(f"Fine-tuned LGBM MAE: {mae_tuned:.2f}")
print(f"Fine-tuned LGBM MAPE: {mape_tuned:.2f}")
```

```python
from xgboost import XGBRegressor
from skopt import BayesSearchCV
from skopt.space import Categorical, Real, Integer
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error

# Define the parameter space for XGBoost using skopt.space types
param_grid_xgb = {
    'n_estimators': Integer(100, 500),
    'learning_rate': Real(0.01, 0.3, prior='log-uniform'),
    'max_depth': Integer(3, 10),
    'subsample': Real(0.5, 1.0, prior='uniform'),
    'colsample_bytree': Real(0.5, 1.0, prior='uniform'),
    'gamma': Real(0, 0.5),
    'reg_alpha': Real(0, 1.0),
    'reg_lambda': Real(0, 1.0)
}

# Initialize XGBRegressor
xgb = XGBRegressor(random_state=42, tree_method='hist') # Use 'hist' for faster training if available

# Initialize BayesSearchCV
bayes_search_xgb = BayesSearchCV(
    estimator=xgb,
    search_spaces=param_grid_xgb,
    scoring='neg_mean_squared_error', # Using negative MSE for maximization
    cv=3, # 3-fold cross-validation
    n_jobs=-1, # Use all available cores
    verbose=1,
    n_iter=50 # Number of iterations for Bayesian Optimization
)

# Fit BayesSearchCV to the training data
bayes_search_xgb.fit(x_train, y_train)

# Print the best parameters and best score
print(f"Best Parameters for XGBoost: {bayes_search_xgb.best_params_}")
print(f"Best Negative MSE for XGBoost: {bayes_search_xgb.best_score_:.2f}")

# Get the best model
best_xgb_model = bayes_search_xgb.best_estimator_

# Make predictions on the test set with the best model
y_pred_tuned_xgb = best_xgb_model.predict(x_test)

# Evaluate the best model
mse_tuned_xgb = mean_squared_error(y_test, y_pred_tuned_xgb)
mae_tuned_xgb = mean_absolute_error(y_test, y_pred_tuned_xgb)
mape_tuned_xgb = mean_absolute_percentage_error(y_test, y_pred_tuned_xgb)
r2_tuned_xgb = r2_score(y_test, y_pred_tuned_xgb)

print(f"\nFine-tuned XGBoost Mean square Error: {mse_tuned_xgb:.2f}")
print(f"Fine-tuned XGBoost R2: {r2_tuned_xgb:.2f}")
print(f"Fine-tuned XGBoost MAE: {mae_tuned_xgb:.2f}")
print(f"Fine-tuned XGBoost MAPE: {mape_tuned_xgb:.2f}")
```
